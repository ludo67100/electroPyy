```python
#First let's create x and y data arrays
import numpy as np 
x = np.array([10,2,5,6,9,8,25,20,12,52,34,18,25,6,17])
y = np.array([10,2,6,7,9,9,35,27,14,60,37,19,26,10,19])

#Let's plot the data first 
from matplotlib import pyplot as plt 
plt.figure()
plt.scatter(x,y,label='data',color='blue')
plt.xlabel('x data')
plt.ylabel('y data')
plt.legend(loc='best')
plt.show();
```

```python
#That look somehow correlated... But let's check it seriously 
#First import electroPyy
import electroPyy

#We have to define the parameters for the regression:
#px: predicted x array / nom: linear regression curve / lpb-upb: lower and upper prediction bands
#r2: squared correlation coefficient / std: standard deviation of y array

px, nom, lpb, upb, r2, std, coeffs = electroPyy.core.Regression.LinReg(x,y,conf=0.95,printparams=True,plot=False) 
```

```python
#Ok it's correlated... But I only trust what I see !! Let's plot it 
plt.figure()
plt.figure()
plt.scatter(x, y, s=20, label='Data')
# plot the regression
plt.plot(px, nom, c='orange', label='y=a x + b',linewidth=2)

# uncertainty lines (95% confidence)
plt.plot(px, nom - 1.96 * std, c='0.5',linestyle='--',\
         label='95% Confidence Region')
plt.plot(px, nom + 1.96 * std, c='0.5',linestyle='--')
# prediction band (95% confidence)
plt.plot(px, lpb, color='0.5',label='95% Prediction Band',linestyle=':')
plt.plot(px, upb, color='0.5',linestyle=':')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(loc='best')
plt.title('Linear Reg : R$^2$={}'.format(round(r2,2)))
plt.show();
```

```python
#Ok, I want to compute a Fourier's transform on my signal. Go. 

#First, the usual 
import numpy as np
import electroPyy
from matplotlib import pyplot as plt 

#We need a signal; let's create one ! How about 10s sampled at 10kHz ?
t0 = 0
t1 = 10
sampling_rate = 10000

#Get the sampling period
sampling_period = 1./sampling_rate

#Points in the signal
N = sampling_rate*(t1-t0)

#Time vector
t = np.linspace(t0,t1,N)

#Now let's do a sinusoid signal with fundamental freq at 2Hz and harmonic at 8Hz
signal = 4*np.sin(2*2*np.pi*t) + 8*np.sin(8*2*np.pi*t)

#Let's plot it to see how beautiful it is 
plt.plot(t,signal)
plt.xlabel('Time(s)');plt.ylabel('Signal Amp.')
plt.show();
```

```python
#How do I get the frequency domain of my signal now ? 
the_fft, freq_domain = electroPyy.core.Convolution.FFT(signal, sampling_rate=sampling_rate,plot=True,freq_range=[0,20])
```
